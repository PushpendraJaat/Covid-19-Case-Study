Q2.2: Generate plots of confirmed cases over time for the top countries.

Finding top countries as per total number of cases.

```python
country_confirmed_cases = covid_19_confirmed.groupby('Country/Region').sum().iloc[:, 4:].sum(axis=1).sort_values(ascending=False)
country_confirmed_cases
```

```python
top_countries = country_confirmed_cases.head(3).index

date_columns = covid_19_confirmed.columns[4:]
covid_19_confirmed[date_columns] = covid_19_confirmed[date_columns].apply(pd.to_numeric, errors='coerce')

# Group data by country and sum values (to remove province/state duplicates)
covid_grouped = covid_19_confirmed.groupby("Country/Region")[date_columns].sum()
date_index = pd.to_datetime(date_columns, format='%m/%d/%y')

for country in top_countries:

    transposed_data = covid_grouped.loc[country]
    transposed_data.index = date_index
    monthly_data = transposed_data.resample('M').sum().diff()

    plt.figure(figsize=(10, 6))
    ax = sns.lineplot(x=monthly_data.index, y=monthly_data.values)

    ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.title(f'Monthly New Confirmed Cases in {country}', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('New Confirmed Cases (Millions)', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f"monthly_new_cases_{country}.png")
    plt.show()
```

**Question 7: Data Merging**

Q7.1: How would you merge the transformed datasets of confirmed cases, deaths, and recoveries on the 'Country/Region' and 'Date' columns to create a comprehensive view of the pandemic's impact?

```python
covid_19_confirmed.head()
```

```python
confirm_cases_long = pd.melt(covid_19_confirmed, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Confirmed')
confirm_cases_long
```

```python
confirm_cases_long['Date'] = pd.to_datetime(confirm_cases_long['Date'])
confirm_cases_long
```

```python
covid_19_recovered.head()
```

```python
recovered_cases_long = pd.melt(covid_19_recovered, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Recovered')
recovered_cases_long['Date'] = pd.to_datetime(recovered_cases_long['Date'])
recovered_cases_long
```

```python
merged_covid_df = pd.merge(confirm_cases_long, recovered_cases_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'], how='outer')
merged_covid_df = pd.merge(merged_covid_df, dealth_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'], how='outer')
merged_covid_df.fillna(0, inplace=True)
merged_covid_df.head(20)
```

```python
merged_covid_df.dtypes
```

```python
merged_covid_df['Lat'] = merged_covid_df['Lat'].astype(object)
merged_covid_df['Long'] = merged_covid_df['Long'].astype(object)
merged_covid_df['Confirmed'] = merged_covid_df['Confirmed'].astype(int)
merged_covid_df['Recovered'] = merged_covid_df['Recovered'].astype(int)
merged_covid_df['Deaths'] = merged_covid_df['Deaths'].astype(int)
```

```python
merged_covid_df.dtypes
```

```python
merged_covid_df
```

```python
merged_covid_df['Active'] = merged_covid_df['Confirmed'] - merged_covid_df['Deaths'] - merged_covid_df['Recovered']
merged_covid_df
```

Q7.2: Analyze the monthly sum of confirmed cases, deaths, and recoveries for countries to understand the progression of the pandemic.[From the merged dataset]

```python
# Step 1: Compute daily differences per Province/State
cols_to_diff = ['Confirmed', 'Deaths', 'Recovered', 'Active']
merged_covid_df[cols_to_diff] = (
    merged_covid_df
    .groupby(['Country/Region', 'Province/State'])[cols_to_diff]
    .diff()
    .fillna(0)
    .clip(lower=0)  # Step 2: Clip negative values in one go
)

# Step 3: Aggregate daily data by Country
country_daily_df = (
    merged_covid_df
    .groupby(['Country/Region', 'Date'])[cols_to_diff]
    .sum()
    .reset_index()
)

# Step 4: Aggregate into monthly totals
monthly_summary = (
    country_daily_df
    .groupby(['Country/Region', country_daily_df['Date'].dt.to_period('M')])[cols_to_diff]
    .sum()
    .reset_index()
    .rename(columns={'Date': 'Month'})
)

# Convert Period to string for readability
monthly_summary['Month'] = monthly_summary['Month'].astype(str)

# Final output
print(monthly_summary.head())
```

```python
india_monthly = monthly_summary[monthly_summary['Country/Region'] == 'India'].copy()

plt.figure(figsize=(14, 6))
for col in ['Confirmed', 'Deaths', 'Recovered', 'Active']:
    plt.plot(india_monthly['Month'].astype(str), india_monthly[col], marker='o', label=col)

plt.title('Monthly COVID-19 Cases in India', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Cases (in Millions)', fontsize=12)
plt.xticks(rotation=45)

# Format Y-axis to show in millions
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f}M'))

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.savefig("monthly_cases_in_India.png")
plt.show()
```

Q8.2: Using the merged dataset, compare the total number of recoveries to the total number of deaths in South Africa. What can this tell us about the outcomes of COVID-19 cases in the country?

```python
south_africa_data = merged_covid_df[merged_covid_df['Country/Region'].str.strip() == 'South Africa']


total_recoveries = int(south_africa_data['Recovered'].sum())
total_deaths = int(south_africa_data['Deaths'].sum())
import matplotlib.pyplot as plt


labels = ['Recoveries', 'Deaths']
sizes = [total_recoveries, total_deaths]
colors = ['green', 'red']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Recoveries vs Deaths in South Africa (2020)', fontsize=14)
plt.axis('equal')  # Ensures the pie is a circle

plt.tight_layout()
plt.savefig("Proportion of Recoveries vs Deaths in South Africa (2020).png")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
bars = plt.barh(['Recoveries', 'Deaths'], [total_recoveries, total_deaths], color=['green', 'red'])

# Add value labels
for bar in bars:
    xval = bar.get_width()
    plt.text(xval + max(total_recoveries, total_deaths)*0.01, bar.get_y() + bar.get_height()/2, f'{xval:,}', va='center')

plt.xlabel('Number of Cases')
# Format x-axis to show in millions
plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f}M'))
plt.title('Total Recoveries vs Deaths in South Africa')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
plt.savefig("Total Recoveries vs Deaths in South Africa.png")
plt.show()
```

Q8.3: Analyze the ratio of recoveries to confirmed cases for the United States monthly from March 2020 to May 2021. Which month experienced the highest recovery ratio, and what could be the potential reasons?

```python
us_monthly = monthly_summary.loc[
    (monthly_summary['Country/Region'] == 'US') &
    (monthly_summary['Month'] >= '2020-03') &
    (monthly_summary['Month'] <= '2021-05')
].copy()

# Convert Month column to datetime for plotting
us_monthly['Month'] = pd.to_datetime(us_monthly['Month'])

# Avoid division by zero
us_monthly = us_monthly[us_monthly['Confirmed'] > 0]

# Calculate recovery ratio
us_monthly['Recovery Ratio'] = us_monthly['Recovered'] / us_monthly['Confirmed']

max_idx = us_monthly['Recovery Ratio'].idxmax()
highest_month = us_monthly.loc[max_idx, 'Month']
highest_ratio = us_monthly.loc[max_idx, 'Recovery Ratio']

print(f"\nMonth with the highest recovery ratio: {highest_month.strftime('%Y-%m')}")
print(f"Highest recovery ratio: {highest_ratio:.2%}")

# Plot
plt.figure(figsize=(12, 6))
ax = plt.gca()
us_monthly.plot(x='Month', y='Recovery Ratio', marker='o', ax=ax, color='green', label='Recovery Ratio')


ax.yaxis.set_major_formatter(mticker.PercentFormatter(1.0))
plt.title('Monthly Recovery Ratio in the US (Mar 2020 - May 2021)', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Recovery Ratio (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.legend()
plt.savefig("Monthly Recovery Ratio in the US (Mar 2020 - May 2021).png")
plt.show()
```
