Q2.2: Generate plots of confirmed cases over time for the top countries.

Finding top countries as per total number of cases.

```python
country_confirmed_cases = covid_19_confirmed.groupby('Country/Region').sum().iloc[:, 4:].sum(axis=1).sort_values(ascending=False)
country_confirmed_cases  #countries with total cases
```

```python
#Convert date columns to numeric (safe conversion)
date_columns = covid_19_confirmed.columns[4:]
covid_19_confirmed[date_columns] = covid_19_confirmed[date_columns].apply(pd.to_numeric, errors='coerce')

#Select top 3 countries
top_countries = country_confirmed_cases.head(3).index

#Group data by country (sum provinces/states)
covid_grouped = covid_19_confirmed.groupby("Country/Region")[date_columns].sum()

# Create a datetime index from the column names
date_index = pd.to_datetime(date_columns, format='%m/%d/%y')

# Plot for each top country
for country in top_countries:
    # Get country data
    country_data = covid_grouped.loc[country]
    country_data.index = date_index

    # Resample monthly and calculate new cases (monthly difference)
    monthly_new_cases = country_data.resample('M').sum().diff()

    # Plotting
    plt.figure(figsize=(10, 6))
    ax = sns.lineplot(x=monthly_new_cases.index, y=monthly_new_cases.values, label=country)

    # Format y-axis as millions
    ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))

    plt.grid(True, linestyle='--', alpha=0.6)
    plt.title(f'Monthly New Confirmed Cases in {country}', fontsize=16)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('New Confirmed Cases (Millions)', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f"../images/monthly_new_cases_{country}.png")
    plt.show()
```

Q7.2: Analyze the monthly sum of confirmed cases, deaths, and recoveries for countries to understand the progression of the pandemic.[From the merged dataset]

```python
# Step 1: Compute daily differences per Province/State
cols_to_diff = ['Confirmed', 'Deaths', 'Recovered', 'Active']
merged_covid_df[cols_to_diff] = (
    merged_covid_df
    .groupby(['Country/Region', 'Province/State'])[cols_to_diff]
    .diff()
    .fillna(0)
    .clip(lower=0)  # Step 2: Clip negative values in one go
)

# Step 3: Aggregate daily data by Country
country_daily_df = (
    merged_covid_df
    .groupby(['Country/Region', 'Date'])[cols_to_diff]
    .sum()
    .reset_index()
)

# Step 4: Aggregate into monthly totals
monthly_summary = (
    country_daily_df
    .groupby(['Country/Region', country_daily_df['Date'].dt.to_period('M')])[cols_to_diff]
    .sum()
    .reset_index()
    .rename(columns={'Date': 'Month'})
)

# Convert Period to string for readability
monthly_summary['Month'] = monthly_summary['Month'].astype(str)

# Final output
print(monthly_summary.head())
```

```python
india_monthly = monthly_summary[monthly_summary['Country/Region'] == 'India'].copy()

plt.figure(figsize=(14, 6))
for col in ['Confirmed', 'Deaths', 'Recovered', 'Active']:
    plt.plot(india_monthly['Month'].astype(str), india_monthly[col], marker='o', label=col)

plt.title('Monthly COVID-19 Cases in India', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Cases (in Millions)', fontsize=12)
plt.xticks(rotation=45)

# Format Y-axis to show in millions
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f}M'))

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.savefig("../images/monthly_cases_in_India.png")
plt.show()
```

Q8.1: For the combined dataset, identify the three countries with the highest average death rates (deaths/confirmed cases) throughout 2020. What might this indicate about the pandemic's impact in these countries?

```python
# Filter data for year 2020
df_2020 = merged_covid_df[merged_covid_df['Date'].dt.year == 2020]

# Group by country and sum confirmed + deaths
country_summary = (
    df_2020.groupby('Country/Region')[['Confirmed', 'Deaths']]
    .sum()
    .query("Confirmed > 0")  # filter directly here
)

# Compute death rate
country_summary['Death Rate'] = country_summary['Deaths'] / country_summary['Confirmed']

# Get top 3 countries by death rate
top_death_rates = country_summary['Death Rate'].nlargest(3)

# Display results
print("Top 3 Countries by Average Death Rate in 2020:\n", top_death_rates)

```

Q8.3: Analyze the ratio of recoveries to confirmed cases for the United States monthly from March 2020 to May 2021. Which month experienced the highest recovery ratio, and what could be the potential reasons?

```python
us_monthly = monthly_summary.loc[
    (monthly_summary['Country/Region'] == 'US') &
    (monthly_summary['Month'] >= '2020-03') &
    (monthly_summary['Month'] <= '2021-05')
].copy()

# Convert Month column to datetime for plotting
us_monthly['Month'] = pd.to_datetime(us_monthly['Month'])

# Avoid division by zero
us_monthly = us_monthly[us_monthly['Confirmed'] > 0]

# Calculate recovery ratio
us_monthly['Recovery Ratio'] = us_monthly['Recovered'] / us_monthly['Confirmed']

max_idx = us_monthly['Recovery Ratio'].idxmax()
highest_month = us_monthly.loc[max_idx, 'Month']
highest_ratio = us_monthly.loc[max_idx, 'Recovery Ratio']

print(f"\nMonth with the highest recovery ratio: {highest_month.strftime('%Y-%m')}")
print(f"Highest recovery ratio: {highest_ratio:.2%}")

# Plot and save img
plt.figure(figsize=(12, 6))
ax = plt.gca()
us_monthly.plot(x='Month', y='Recovery Ratio', marker='o', ax=ax, color='green', label='Recovery Ratio')


ax.yaxis.set_major_formatter(mticker.PercentFormatter(1.0))
plt.title('Monthly Recovery Ratio in the US (Mar 2020 - May 2021)', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Recovery Ratio (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.legend()
plt.savefig("../images/Monthly Recovery Ratio in the US (Mar 2020 - May 2021).png")
plt.show()
```
